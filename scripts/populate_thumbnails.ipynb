### thumbnails

```python
import pandas as pd
```

```python
# file paths
primary_csv_path = r"inputs\rutgers-primary-2026-07-05-163400.csv"

distributions_csv_path = r"inputs\rutgers-distributions-2026-07-05-163423.csv"

output_csv_path = r"outputs\rutgers_primary_thumbnails.csv"
```

```python
# load CSVs into data frames

primary_df = pd.read_csv(primary_csv_path)

dist_df = pd.read_csv(distributions_csv_path)
```

```python
# Filter distributions to only JPEG rows
jpeg_df = dist_df[dist_df["label"] == "JPEG"]
```

```python
# Keep only relevant columns
jpeg_df = jpeg_df[["friendlier_id", "distribution_url"]]
```

```python
# de-dupe

jpeg_df = jpeg_df.drop_duplicates(subset="friendlier_id")
```

```python
# Merge onto primary
merged_df = primary_df.merge(
    jpeg_df,
    how="left",
    left_on="ID",
    right_on="friendlier_id"
)
```

```python
# Populate "B1G Image" field
merged_df["B1G Image"] = merged_df["distribution_url"]
```

```python
# Drop helper columns
merged_df = merged_df.drop(columns=["friendlier_id", "distribution_url"])
```

```python
# Save result
merged_df.to_csv(output_csv_path, index=False)

print("Done! Output saved to:", output_csv_path)
```
